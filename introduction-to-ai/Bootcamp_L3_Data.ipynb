## 2. Understanding data as points in space

### Data points

In order to see the data we'll have to understand the view it from a geometric perspective.

Let's firstly use the basic dataset with houses and we will then come back to the houses. For the moment, we need only the numerical values, so we are going to ignore the columns that are text.

```python
house_data
```

We will now select two columns: Squared Meters and Price and plot them. Each house will be a point on the graph.

On the **horizontal axis**, we will represent the **Squared Meters** and on the *vertical axis* we will have the *Price*. Let's see!

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

icon_path = 'house_icon.png'
icon = mpimg.imread(icon_path)

def add_icon(ax, x, y, icon, zoom=0.04):
    imagebox = OffsetImage(icon, zoom=zoom)
    ab = AnnotationBbox(imagebox, (x, y), frameon=False)
    ax.add_artist(ab)

fig, ax = plt.subplots()

# this is the command for plotting the points: the first argument is an array with values on horizontal axis, the second arg is an array with values on vertical axis
ax.scatter(house_data['Squared Meters'], house_data['Price'], color='blue')

# this is for plotting number above the points
for i in range(len(house_data)):
    plt.annotate(i, (house_data['Squared Meters'][i], house_data['Price'][i]), textcoords="offset points", xytext=(0,12), ha='center')

for i in range(len(house_data)):
    add_icon(ax, house_data['Squared Meters'][i], house_data['Price'][i], icon)

plt.title("Houses (Price vs Squared Meters)")
plt.xlim((20, 300))
plt.ylim((90000, 750000))
plt.grid()
plt.xlabel("Squared Meters")
plt.ylabel("Price (in $)")

plt.show()
```

```python
house_data
```

Alright, each blue point corresponds to a sample. More precisely, each point is a house. Furthermore, by plotting the points we can see a trend. What can we visually infer from the plot?

Exactly, the bigger the space, the higher the price.

This is a 2-dimensional space so it depicts the relationship between two features. Let's take it to the next level and also include the rooms. Now we're going to a 3-dimensional space. For that we will use plotly which is used for **interactive** plots. These are cooler... you can play with the space and see the points from multiple angles.

Distance between points in space is crucial. We instantly perceived the points and compute the distances. But how to compute them using computers?

We can use **Euclidean distances** between two points to check if they are close or far.

The closer they are, the more similar they are whereas when they are far from each other, it means they are different.

## 3. Data as vectors

```python
house_data.head()
```

Previously, we extracted just a column using a similar syntax to dicts. Let's extract a row which represents a single sample.

```python
house0_features = house_data.iloc[0] # iloc means Integer LOCation
house0_features
```

Let's remove the header from the previous view and focus only on the numerical attributes.

```python
list(house0_features[:4])
```

A single sample can be described as a **list** | **vector** | **array** of its features.

Vector is a term that is use in multiple fields such as maths, physics and computer science. At its core, it refers to an entity described by a list of numbers.


In computer science, we use the term vector to denote a fixed-length sequence. Usually, the order in a vector is very important. Look at the above example and imagine changing the last element with the first. It will mean the price is 3$ and there are 400000 rooms. That will be a crucial mistake. Regarding the length, we also refer to it as size or dimension.



In maths and physics, vectors are important concepts as well because they have *magnitude* and *direction* ðŸ˜Ÿ.

Imagine you are passing a ball to your teammate. You have to pass it with a certain power (magnitude) and in a certain direction. Now imagine you are on a field. You are point A (the red one) and he is point B (the blue one). The ball is going from A to B and it goes for a certain distance. The vector will show the path of the ball having also an arrow on the head to emphasize the direction.


```python
# the code is not important to understand -> just for the visualization

import matplotlib.pyplot as plt
import numpy as np

point_A = np.array([2, 2])
point_B = np.array([6, 8])

plt.figure(figsize=(6, 6))
plt.quiver(point_A[0], point_A[1],
           point_B[0] - point_A[0], point_B[1] - point_A[1],
           angles='xy', scale_units='xy', scale=1, color='purple', label="Vector AB")

# Mark points A and B
plt.scatter(*point_A, color='red', label="Point A (2,2)", zorder=5)
plt.scatter(*point_B, color='blue', label="Point B (6,8)", zorder=5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Vector from A to B")
plt.legend()
plt.grid(True)
plt.axis('equal')

plt.xlim((-5, 10))
plt.ylim((-5, 10))
plt.show()
```


Magnitude and direction are present explicitly in computer science mostly in games or graphics.

Alright, that's enough... back to our data!

Are these geometric vectors really important to us? Are magnitude and direction relevant to us? Why don't we just treat the data points as simple points and that's it?

**ANSWER:** No, for the moment magnitude and direction is not relevant to us. Despite this, **in some cases** we will treat our data points as geometric vectors because both *geometric* vectors and our *vectors* (lists of numbers) have two properties that behave exactly the same: **addition** and **multiplication**.

Imagine you own house0 and then you buy house1. For more intuition, imagine that both of them are part of a duplex. Now you own both the houses - you own the entire duplex. If you have to describe your entire duplex as a new single vector including price, rooms, squared meters, bathrooms, how would you do that?

```python
house_data.head(1)
```

```python
house0 = list(house_data.iloc[0][:4])
house1 = list(house_data.iloc[1][:4])

house0, house1
```

Exactly, you sum component by component.

Your duplex is represented by a new vector computed as a sum of the two houses.

```python
duplex = house0 + house1
duplex
```

Oh my god. That's bad. There should be another way.

```python
# zip(house0, house1) => [(400000, 150000), (130, 50), (4, 2), (3, 1)]
```

```python
duplex = []
for x, y in zip(house0, house1):
    duplex.append(x + y)
duplex
```

Ok, now it looks good. My duplex is worth 55000, it has a total of 180 sq m, 6 rooms and 4 bathrooms. Cool!

But the code is just too much for a simple sum between two vectors. Here is where numpy saves the day. Numpy is a very efficient library designed for this vectors.

```python
import numpy as np

house0 = np.array(house_data.iloc[0][:4], dtype=int) # create an array - but a special one a numpy array with ints
house1 = np.array(house_data.iloc[1][:4], dtype=int)

house0, house1
```

```python
# [1, 2, 3, 4]
# array([1, 2, 3, 4]) # un pic mai mult mai inteligent si mai eficient
```

We can recognize it because it has a specific mark: array([...]) instead of simply [...].

```python
duplex = house0 + house1
duplex
```

Ok. That's much better.

Alright, now imagine you buy 4 houses in a fourplex, all of them similar to house2. How you will describe now your entire fourplex?

Exactly, you will multiply by 4 on each column.

```python
house2 = list(house_data.iloc[2][:4])

# v = [1, 2, 3]
# 3 * v = [3, 6, 9]

fourplex = 4 * house2
fourplex
```

Ok, that's bad. Let's try again with numpy.

```python
house2 = np.array(house_data.iloc[2][:4], dtype=int)

fourplex = 4 * house2
fourplex
```

Ok, better.

**CONCLUSION:** Geometric vectors have this exact same properties. When you add two vectors, you add component by component.
When you multiply a vector with a scalar, you multiply each component by that scalar.
